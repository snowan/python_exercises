```python
## Add blur and save image
def apply_blur(image_name, newname, k):
    image = cv2.imread(image_name)
    blurred = cv2.blur(image, (k,k)) 
    cv2.imwrite(newname, blurred)
       
```

```python
# crop image
def crop(image_name, size, newname):
    image = cv2.imread(image_name)
    width, height = image.size   # Get dimensions
    new_width = size
    new_height = size

    left = (width - new_width)/2
    top = (height - new_height)/2
    right = (width + new_width)/2
    bottom = (height + new_height)/2

    new_img = image.crop((left, top, right, bottom))
    cv2.imwrite(newname, new_img)
```

```python
# resize image and save
def resize_image(image_name, newname, width, height):
    image = cv2.imread(image_name)
    new_img = cv2.resize(image, (width,height))
    cv2.imwrite(newname, new_img)
```

```python
# translation and save image
def tranlation_img(image_name, newname):
    img = cv2.imread(image_name)
    print img.shape
    rows,cols = img.shape

    M = np.float32([[1,0,100],[0,1,50]])
    new_img = cv2.warpAffine(img,M,(cols,rows))
    cv2.write(newname, new_img)
```

```python
def rotate(image_name, newname, angle, center = None, scale = 1.0):
    image = cv2.imread(image_name)
    (h, w) = image.shape[:2]

    if center is None:
        center = (w / 2, h / 2)

    # Perform the rotation
    M = cv2.getRotationMatrix2D(center, angle, scale)
    rotated = cv2.warpAffine(image, M, (w, h))
    cv2.imwrite(newname, rotated)
```
